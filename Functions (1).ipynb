1.	What is the difference between a function and a method in Python?
Ans- In Python, functions and methods are both used to perform actions or computations, but they differ mainly in how they are called and associated with data.
Function- Function is a reusable code that is not associated with an object.it is called using the def keyword and can be Called directly using its name.


```python
#FUNCTION
def greet(name):
    return f"Hello, {name}!"

print(greet("Anmol"))
#METHOD
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Anmol"))
```

 Explain the concept of generators in Python and how they are defined.

 A generator is a special kind of iterator in Python that allows you to generate values on the fly using the yield keyword, instead of creating and storing the entire sequence in memory. Or a generator is a function that returns an iterator and produces a sequence of values one at a time, using the yield statement.

Generators are defined using regular Python functions, but instead of return, they use yield.


```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(4):
    print(num, end= " ")
```

What are the advantages of using generators over regular functions?

Generators offer several important benefits over regular functions, especially when dealing with large data, performance-sensitive tasks.

1.Memory Efficiency-- Generators do not store the entire data in memory. They yield one item at a time, which is ideal when working with large datasets.

2.Improved Performance--Generators start execution immediately without waiting to build the full data structure. This can speed up processes where only part of the data is used.

3.Infinite Sequences Support-- Generators are perfect for defining sequences with no fixed end, such as streaming live data or generating Fibonacci numbers.


```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

for num in fibonacci():
    if num > 100:
        break
    print(num)
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.
def sum_of_even(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
numbers = [10, 15, 20, 25, 30]
print(sum_of_even(numbers))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from
# 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):  # check all numbers from 2 to n-1
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given
# exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in power_of_two(5):
    print(power)
```
